```python
# Import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ---Load the dataset and Inspection----
# Load the dataset
file_name = 'Smartphones_cleaned_dataset.csv'
df = pd.read_csv(file_name)
```

```python
# Display the first 5 rows of the DataFrame
print("\nInformation about the DataFrame:")
print(df.info())
```

```python
# Section 1. --- Data Cleaning and Preprocessing ---
# Check for duplicates and drop them if any
print("\nNumber of duplicates rows:", df.duplicated().sum())
df.drop_duplicates(inplace=True)

#Drop rows with missing values for simplicity in EDA
df.dropna(inplace=True)
```

```python
# Display the updated DatFrame info
print("\nDataFrame Info after dropping missing values:")
print(df.info())
```

```python
# Section 3. ---Exploratory Data Analysis (EDA)---
# Calculate descriptive statistics for key numerical columns
numerical_cols = ['price', 'rating', 'ram_capacity', 'internal_memory', 'screen_size', 'refresh_rate', 'battery_capacity', 'primary_camera_rear', 'primary_camera_front', 'resolution_width', 'resolution_height']
print("\nDescriptive statistics for numerical columns:")
print(df[numerical_cols].describe().T.to_markdown(numalign="left", stralign="left"))
```

```python
# Find value counts for categorical columns
print("\nValue counts for 'brand_name' (top 10):")
print(df['brand_name'].value_counts().head(10).to_markdown(numalign="left", stralign = "left"))

print("\nValue counts for 'brand_name' (top 10):")
print(df['brand_name'].value_counts().head(10).to_markdown(numalign = "left" , stralign ="left"))

print("\nValue counts for 'processor_brand':")
print(df['processor_brand'].value_counts().head(10).to_markdown(numalign ="left" , stralign= "left"))

print("\nValue counts for 'has_5g':")
print(df['has_5g'].value_counts().head(10).to_markdown(numalign ="left" , stralign= "left"))

print("\nValue counts for 'os':")
print(df['os'].value_counts().head(10).to_markdown(numalign = "left" , stralign ="left"))
```

```python
# Section 5: --- Visualization---
# Create a histogram for the price distribution 
plt.figure(figsize=(10,6))
sns.histplot(df['price'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Smartphone Price (INR)', fontsize=16)
plt.xlabel('Price (INR)',fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('price_distribution.png')
```

```python
# Create a bar chart for the top 10 brands
top_brands =df['brand_name'].value_counts().head(10).sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_brands.index, y=top_brands.values, palette='viridis')
plt.title('Top 10 Most Common Smartphone Brands', fontsize=16)
plt.xlabel('Brand Name', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('top_10_brands.png')
```

```python
# Create a boxplot for price by processor brand 
plt.figure(figsize=(12,7))
sns.boxplot(x='processor_brand', y='price', data=df, order=df['processor_brand'].value_counts().index, palette='Set2')
plt.title('Price Distribution by Processor Brand', fontsize=16)
plt.xlabel('Processor Brand', fontsize=12)
plt.ylabel('Price (INR)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('price_by_processor_brand.png')
```

```python
# Create a bar chart for 5g availability
has_5g_counts =df['has_5g'].value_counts()
plt.figure(figsize=(8,6))
sns.barplot(x=has_5g_counts.index, y=has_5g_counts.values, palette='cividis')
plt.title('Distribution of 5G Availability', fontsize=16)
plt.xlabel('Has 5G', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(ticks=[0,1], labels=['No (False)', 'Yes (True)'])
plt.tight_layout()
plt.savefig('5g_availability.png')
```

```python
# Create a scatter plot of price vs. rating
plt.figure(figsize=(10, 8))
sns.scatterplot(x='rating', y='price', data=df, alpha=0.6, color='darkred')
plt.title('Relationship between Price and Rating', fontsize=16)
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Price (INR)', fontsize=12)
plt.grid(linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('price_vs_rating.png')
```
