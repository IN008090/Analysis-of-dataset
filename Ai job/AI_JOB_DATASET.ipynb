```python
# Section 1. Load and Inspect the data
import pandas as pd
 #load the dataset
df= pd.read_csv('ai_job_dataset.csv')
```

```python
#Display the first 5 rwos of the DataFrame
print(" ---First 5 rows of the DataFrame--- ")
print(df.head().to_markdown(index=False, numalign="left", stralign="left"))
```

```python
# Display information about the DataFrame
print("\nInformation about the DataFrame:")
print(df.info())
```

```python
# Section 2: Descriptive Statistics
# Calculate descriptive statistics for numerical columns
numerical_cols = ['salary_usd', 'remote_ratio', 'years_experience', 'job_description_length', 'benefits_score']
print("Descriptive statistics for numerical columns:")
print(df[numerical_cols].describe().to_markdown(numalign="left", stralign="left"))
```

```python
# Section 3. Categorical Data Analysis
# Find value counts for categorical columns
print("Value counts for `job_title`:")
print(df['job_title'].value_counts().head(10).to_markdown(numalign="left", stralign="left"))

print("\nValue counts for `experience_level`:")
print(df['experience_level'].value_counts().to_markdown(numalign="left", stralign="left"))

print("\nValue counts for `employment_type`:")
print(df['employment_type'].value_counts().to_markdown(numalign="left", stralign="left"))

print("\nValue counts for `education_required`:")
print(df['education_required'].value_counts().to_markdown(numalign="left", stralign="left"))
```

```python
# Section 4: Visualization 
import matplotlib.pyplot as plt
import seaborn as sns
# Create a bar chart for the top 10 job titles
top_job_titles = df['job_title'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_job_titles.values, y=top_job_titles.index, palette='viridis')
plt.title('Top 10 AI Job Titles', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Job Title', fontsize=12)
plt.tight_layout()
plt.savefig('top_10_job_titles.png')
```

```python
# Create a histogram for the salary_usd distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['salary_usd'], bins=30, kde=True, color='skyblue')
plt.title('Distrubution of Salary (in USD)', fontsize=16)
plt.xlabel('Salary (USD)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('salary_distribution.png')
```
